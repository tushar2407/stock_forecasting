```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
```

```python
sp500
```

```python
sp500.plot.line(y='Close', title='S&P 500 Close Price', ylabel='Close Price (USD)')
```

```python
sp500 = sp500[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
sp500.loc[:,'Tomorrow Close'] = sp500['Close'].shift(-1)
sp500
```

```python
sp500['Target'] = (sp500['Tomorrow Close'] > sp500['Close']).astype(int)
sp500
```

```python
sp500 = sp500.loc['1990-01-01':].copy()
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = sp500.iloc[:-1000]
test = sp500.iloc[-1000:]

predictors = ['Open', 'High', 'Low', 'Close', 'Volume']
model.fit(train[predictors], train['Target'])
```

```python
test
```

```python
from sklearn.metrics import precision_score

test_predictions = model.predict(test[predictors])
test_predictions
```

```python
import pandas as pd
test_predictions = pd.Series(test_predictions, index=test.index)
```

```python
precision_score(test['Target'], test_predictions)
```

```python
combined = pd.concat([test['Target'], test_predictions], axis=1)
combined.plot()
```

# Building Backtest

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    test_predictions = model.predict(test[predictors])
    test_predictions = pd.Series(test_predictions, index=test.index, name='Predictions')
    combined = pd.concat([test['Target'], test_predictions], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    results = []
    for i in range(start, len(data), step):
        train = data.iloc[:i].copy()
        test = data.iloc[i:i+step].copy()
        predictions= predict(train, test, predictors, model)
        # precision = precision_score(test['Target'], combined.iloc[:,1])
        results.append(predictions)
    return pd.concat(results)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```

## Add more predictors

```python
horizons = [2, 5, 60, 250, 1000 ]
new_predictors = []

for horizon in horizons:
    rolling_average = sp500['Close'].rolling(window=horizon).mean()
    sp500[f'Close_ratio_{horizon}'] = sp500['Close'] / rolling_average

    sp500[f'Trend_{horizon}'] = sp500['Target'].shift(1).rolling(window=horizon).sum()
    
    new_predictors.extend([f'Close_ratio_{horizon}', f'Trend_{horizon}'])
    
```

```python
sp500
```

```python
sp500 = sp500.dropna()
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    test_predictions = model.predict_proba(test[predictors])[:, 1]
    test_predictions[test_predictions >= 0.6] = 1
    test_predictions[test_predictions < 0.6] = 0
    test_predictions = pd.Series(test_predictions, index=test.index, name='Predictions')
    combined = pd.concat([test['Target'], test_predictions], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```
